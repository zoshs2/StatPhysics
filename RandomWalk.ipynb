```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
from datetime import timedelta
```

```python
# 랜덤워크 시드설정
np.random.seed(5)
```

```python
# 1분 단위 시계열 데이터(Time series dataset) 생성
init_time = datetime.datetime(2022, 2, 1, 21, 0, 0) # 2022년 2월 1일 21시 0분 0초부터
time_series = np.array([init_time])
for _ in range(120): # 앞으로 2시간 time_series 생성
    new_time = time_series[-1] + timedelta(minutes=1) # interval of 1 minute
    time_series = np.append(time_series, new_time)

print(time_series[0])
print(time_series[-1])
```

## np.random 모듈의 3가지 대표 메소드
1. np.random.randint(n): 0~(n-1) 사이의 랜덤한 정수 숫자(Uniform dist.로부터 추출); low & high bound 두고 싶으면, np.random.randint(1, 6): 1부터 5사이
<br><br>
2. np.random.rand(m, n): 0부터 1사이 [0, 1) 의 '랜덤한 float형 숫자' m by n matrix 형태로 추출 // 단지 난수 하나 뽑고 싶으면 m,n 입력안하면 됨 np.random.rand()
<br><br>
3. np.random.randn(m, n): 표준정규분포(Standard Normal Distribution; 평균=0, 분산=1)로부터 m by n matrix 형태의 난수 array 추출 // 역시 np.random.randn() 가능
<br>np.random.randn() 는 기본적으로 평균0, 분산1(그럼 당연히 표준편차도 1)인 표준정규분포에서 난수를 추출해오는데, 이를 조절하고 싶으면 <br>
mean + sigma * np.random.randn(...)해서 N(mean, sigma^2)인 정규분포에서 추출하면 된다. 


```python
InitPos = 10000
StepSize = 100
RandomTrail = np.array([InitPos])
for _ in range(120):
    NextPos = RandomTrail[-1] + StepSize * np.around(np.random.randn())
    RandomTrail = np.append(RandomTrail, NextPos)

print(RandomTrail)
```

```python
fig, ax = plt.subplots(figsize=(15,7))
ax.plot(time_series, RandomTrail)
# ax.tick_params(labelrotation=45) # x y 축 ticklabels 모두 회전
# xtick만 회전하고 싶으면
for tick in ax.get_xticklabels():
    tick.set_rotation(45)
plt.tight_layout()
plt.show()
```

# Random Walk 200회 실시

```python
fig, ax = plt.subplots(figsize=(15,7))
for _ in range(200):
    InitPos = 10000
    StepSize = 100
    RandomTrail = np.array([InitPos])
    for _ in range(120):
        NextPos = RandomTrail[-1] + StepSize * np.around(np.random.randn())
        RandomTrail = np.append(RandomTrail, NextPos)
    
    ax.plot(time_series, RandomTrail, alpha=0.5)

for tick in ax.get_xticklabels():
    tick.set_rotation(45)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
for _ in range(200):
    InitPos = 10000
    StepSize = 100
    RandomTrail = np.array([InitPos])
    for _ in range(120):
        NextPos = RandomTrail[-1] + StepSize * np.around(0.5 + (1.5*np.random.randn())) # N(0.5, 1.5**2) 정규분포에서 난수추출
        RandomTrail = np.append(RandomTrail, NextPos)
    
    ax.plot(time_series, RandomTrail, alpha=0.5)

for tick in ax.get_xticklabels():
    tick.set_rotation(45)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
for _ in range(200):
    InitPos = 10000
    StepSize = 100
    RandomTrail = np.array([InitPos])
    for _ in range(120):
        NextPos = RandomTrail[-1] + StepSize * np.around(1.5 + (1.5*np.random.randn())) # N(1.5, 1.5**2) 정규분포에서 난수추출
        RandomTrail = np.append(RandomTrail, NextPos)
    
    ax.plot(time_series, RandomTrail, alpha=0.5)

for tick in ax.get_xticklabels():
    tick.set_rotation(45)

plt.tight_layout()
plt.show()
```

## 아무래도 평균(mean) sign 방향으로 점점 멀어진다.

```python
fig, ax = plt.subplots(figsize=(15,7))
for _ in range(200):
    InitPos = 10000
    StepSize = 100
    RandomTrail = np.array([InitPos])
    for _ in range(120):
        NextPos = RandomTrail[-1] + StepSize * np.around(-.5 + (1.5*np.random.randn())) # N(-0.5, 1.5**2) 정규분포에서 난수추출
        RandomTrail = np.append(RandomTrail, NextPos)
    
    ax.plot(time_series, RandomTrail, alpha=0.5)

for tick in ax.get_xticklabels():
    tick.set_rotation(45)

plt.tight_layout()
plt.show()
```
